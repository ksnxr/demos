```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
def plot(func, yaxis, name):
    plt.ylim(yaxis)
    plt.locator_params(nbins=5)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline(lw=1, c='black')
    plt.axvline(lw=1, c='black')
    plt.grid(alpha=0.4, ls='-.')
    plt.box(on=None)
    plt.plot(x, func(x), c='r', lw=3)
    plt.savefig(name, dpi=100)
```

## Sigmoid

```python
sigmoid = np.vectorize(lambda x: 1/(1+math.exp(-x)))
plot(sigmoid, (-0.4, 1.4), "Sigmoid")
```

## tanh

```python
tanh = np.vectorize(lambda x: (1-math.exp(-2*x))/(1+math.exp(-2*x)))
plot(tanh, (-1.4, 1.4), "tanh")
```

## ReLu

```python
relu = np.vectorize(lambda x: x if x >=0 else 0, otypes=[np.float])
plot(relu, (-0.6, 5.4), "ReLu")
```

## Leaky_ReLu

```python
leaky_relu = np.vectorize(lambda x: x if x >=0 else 0.1*x, otypes=[np.float])
plot(leaky_relu, (-0.6, 5.4), "Leaky_ReLu")
```
